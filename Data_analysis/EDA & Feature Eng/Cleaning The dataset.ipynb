**Algerian Forest Fires Dataset**

Algerian Forest Fires Data from 2012

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class)
The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/Algerian_forest_fires_dataset.csv")
```

```python
dataset.head()
```

**Attributes Information:**

1.Date:(DD/MM/YYYY):('June'to'September'),year(2012)WeatherData.

2.Temp:Temperature(in Celsius degree:22 to 42).

3.RH:Relative Humidity in %21 to 90.

4.Ws:Wind Speed in Km/h:6 to 29.

5.Rain:total day in mm: 0 to 16.8.

6.FFMC:Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5.

7.DMC:Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9.

8.DC:Drought Code (DC) index from the FWI system: 7 to 220.4.

9.inital spreadsheet index:(ISI) index from the FWI system:0 to 18.5.

10.Buildup index(BUI):index from the FWI system:1.1 to 68.

11.Fire Weather index(FWI):0 to 31.1.

12.Classes:two classes,namely Fire and Not Fire.
